```python
from mobbi import Mobbi
from selenium.webdriver.common.by import By
import pandas as pd
from sqlalchemy import create_engine
import time
```

```python
columns = ['car_brand',
           'car_model', 
           'car_variant', 
           'car_price', 
           'car_transmission', 
           'car_mileage',
           'car_year',
           'car_location',
           'car_color',
           'car_engine_capacity',
           'car_fuel'
           ]
df = pd.DataFrame(columns=columns)
```

```python
mobbi = Mobbi(teardown=False)

mobbi.land_first_page()
mobbi.close_popup()
mobbi.click_textbox()
for i in [1, 2, 3]:
    mobbi.select_brand(i)
    num_cars = mobbi.get_number_of_cars()
    for j in range(num_cars):
        while True:
            try:
                car_page = mobbi.car_page(j+1)
                break
            except:
                last_height = mobbi.scroll_page(600)
                time.sleep(3)
        if car_page == False:
            continue
        while True:
            try:
                car_page.click()
                print("hallo")
                break
            except:
                last_height = mobbi.scroll_page(600)
                time.sleep(3)
        car_info = mobbi.get_car_information()
        if car_info == None:
            mobbi.go_back()
            continue
        df.loc[len(df)] = car_info
        mobbi.go_back()
```
